# Correlation of each column

```python


# Make a copy and encode categorical columns
df_encoded = df.copy()
label_encoders = {}

for col in df_encoded.columns:
    if df_encoded[col].dtype == 'object' or df_encoded[col].dtype.name == 'category':
        le = LabelEncoder()
        df_encoded[col] = le.fit_transform(df_encoded[col].astype(str))
        label_encoders[col] = le  # For possible reverse-transform

# Compute correlation matrix
corr_matrix = df_encoded.corr(method='pearson')

# Plot the heatmap
plt.figure(figsize=(18, 10))
sns.heatmap(
    corr_matrix,
    annot=True,
    fmt=".3f",
    cmap="coolwarm",
    center=0,
    linewidths=0.5,
    cbar_kws={"shrink": 0.8}
)

plt.title("Correlation Heatmap", fontsize=16)
plt.xticks(rotation=90)
plt.yticks(rotation=0)
plt.tight_layout()
plt.show()
```

* 1. Columns with high positive correlation 

* Gender & Subscription_status
* Gender & Discount-applied
* Gender & Promo_code_used
* Subscription_status & Discount-applied
* Subscription_status & Promo_code_used
* Discount-applied & Promo_code_used (1.00)

2. Columns with high negative correlation 
* Customer_id & Discount-applied
* Customer_id & Subscription_status
* Customer_id & Promo_code_used
* Customer_id & Gender

4.Unrelated colums (0.000)
 * season & Category
 * size & Location
 * Color & Season
 * Reviw_rating & Season
 * Reviw_rating & Category
 * Color & Shipping_type
 * Color & Previous_purchase

* Customer ID has a strong negative correlation with behavioral features,but Customer ID is a unique identifier, not a real feature.so we can drop Customer ID.

## Age vs Size

```python
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='Age_Group', hue='Size')
plt.xlabel('Age Group')
plt.ylabel('Count')
plt.title('Distribution of Sizes in Each Age Group')
plt.show()
```

```python


age_groups = df['Age_Group'].unique()
age_groups = [ag for ag in age_groups if pd.notnull(ag)]

fig, axes = plt.subplots(1, len(age_groups), figsize=(16, 5))

for i, ag in enumerate(age_groups):
    sizes = df[df['Age_Group'] == ag]['Size'].value_counts()
    axes[i].pie(
        sizes,
        labels=sizes.index,
        autopct='%1.0f%%',
        startangle=90
    )
    axes[i].set_title(f'Age Group: {ag}')
    axes[i].set_ylabel('')

plt.tight_layout()
plt.show()
```


* Across all age groups, the M size is the most purchased.
* M > L > S > XL

* The proportions for each size are very similar in every age group, indicating that age does not significantly affect size preference.

## Insight

* Focus inventory on "M" and "L" sizes for all age groups.
* Less stock is needed for "S" and "XL" sizes.
* No need for age-specific size.

## Age vs Season


* Each age group, with only small differences. This means customers shop throughout the year, regardless of age.

* No age group shows a dramatic spike or drop in any particular season. For example, the 18–25 group has slightly higher purchases in Winter and Spring, but the difference is not large.

* 18–25: Slightly more purchases in Winter and Spring.
* 26–35: Purchases are very balanced across all seasons.
*  36–45: Slightly more in Spring and Fall.
*  46+: Slightly more in Spring and Fall, less in Winter.

## Insights
* Do not need to focus seasonal marketing on a specific age group.
* Spread evenly across seasons for all age groups.
* Consider a small push for younger customers in Winter and Spring.
